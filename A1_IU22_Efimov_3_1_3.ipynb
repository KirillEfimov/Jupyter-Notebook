```python
import os

#os.chdir('..//..//Assignments-master//data//A1_Descriptive_Analysis//stock')
data_folder = '..//..//Assignments-master//data//A1_Descriptive_Analysis//stock'
stock_list = os.listdir(data_folder)

df = pd.DataFrame()

for file_name in stock_list:
        df[file_name[0:-4]] = pd.read_csv(data_folder + '/' + file_name,
                                    usecols=['Date', 'Close'], index_col=0,
                                    parse_dates=True).Close
df
```

```python
corr_df = df.corr(method="pearson")

max_corr = corr_df.GOOGL.drop("GOOGL").idxmax()
min_corr = corr_df.GOOGL.drop("GOOGL").idxmin()
no_corr = abs(corr_df.GOOGL.drop("GOOGL")).idxmin()

print("Акция с максимальной корреляцией: ", max_corr, "\nАкция с минимальной корреляцией: ",
      min_corr, "\nАкция с отсутствием корреляции: ", no_corr)
```

```python
plt.figure(figsize=[15, 5])

plt.subplot(1,3,1)
plt.title("GOOGL " + max_corr)
plt.scatter(df.loc[:, "GOOGL"].tolist(), df.loc[:, max_corr].tolist())
plt.grid(True)

plt.subplot(1,3,2)
plt.title("GOOGL " + min_corr)
plt.scatter(df.loc[:, "GOOGL"].tolist(), df.loc[:, min_corr].tolist())
plt.grid(True)

plt.subplot(1,3,3)
plt.title("GOOGL " + no_corr)
plt.scatter(df.loc[:, "GOOGL"].tolist(), df.loc[:, no_corr].tolist())
plt.grid(True)

plt.tight_layout()

plt.show()
```
